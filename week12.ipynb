```python
import csv
f = open('seoul.csv','r',encoding = 'cp949')
data = csv.reader(f, delimiter = ',')

#index 0:날짜 1:지점 2:평균기온 3:최저기온 4:최고기온
header = next(data)

maxDC = -999   #DailyCross
maxDCdate = ""
minT = 999
minTdate = ""
maxT = -999
maxTdate = ""

overlapMaxT =-999
overlapMinT =999
overlapMaxDC =-999

overlapMaxTDCdate = ""
overlapMinTdate=""
overlapMaxTdate=""

for row in data:
    tempMinT = float(row[3])
    tempMaxT = float(row[4])
    tempMaxDC = tempMaxT - tempMinT
    
    #elif문은 해당data가 동일한경우를 핸들링한다
    #최저기온
    if(minT > tempMinT):   
        minT = tempMinT
        minTdate = row[0]
    elif(minT == tempMinT):
        overlapMinTdate = row[0]
        overlapMinT = tempMinT
    #최고기온   
    if(maxT < tempMaxT):   
        maxT = tempMaxT
        maxTdate = row[0]
    elif(maxT == tempMaxT):
        overlapMaxTdate = row[0]
        overlapMaxT = tempMaxT
    #최고일교차  
    if(maxDC < tempMaxDC):
        maxDC = tempMaxDC
        maxDCdate = row[0]
    elif(maxDC == tempMaxDC):
        overlapMaxDCdate = row[0]
        overlapMaxDC = tempMaxDC
        
        
if(overlapMinT == minT):
    minTdate = minTdate + " 그리고 " +overlapMinTdate
if(overlapMaxT == maxT):
    maxTdate = maxTdate + " 그리고 "+overlapMaxTdate
if(overlapMaxDC == maxDC):
    maxDCdate = maxDCdate + " 그리고 " +overlapMaxDCdate
    
    
print("일교차가 가장 큰 날")
print(maxDCdate)
print("최저기온이 가장 낮은 날")
print(minTdate)
print("최고기온이 가장 높은 날")
print(maxTdate)






f.close()
```
